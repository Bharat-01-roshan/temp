```python
# # import numpy as np
# # import librosa

# def extract_mfcc_delta_features(file_name, n_mfcc=40):
#     # Load audio with fast resampling
#     audio, sample_rate = librosa.load(file_name, res_type='kaiser_fast')

#     # Compute MFCC features
#     mfcc = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=n_mfcc)
#     delta = librosa.feature.delta(mfcc, width=7)

#     # Compute first-order delta (rate of change)
#     # delta = librosa.feature.delta(mfcc)

#     # Compute second-order delta (acceleration)
#     delta2 = librosa.feature.delta(mfcc, order=2)

#     # Stack them horizontally: MFCC + MFCC + MFCC
#     combined = np.hstack((
#         np.mean(mfcc.T, axis=0),
#         np.mean(delta.T, axis=0),
#         np.mean(delta2.T, axis=0)
#     ))

#     return combined

def extract_mfcc_delta_features(file_path, n_mfcc=40, min_frames=9):
    try:
        # Load audio file
        audio, sample_rate = librosa.load(file_path, sr=None)

        # Extract MFCC features
        mfcc = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=n_mfcc)

        # Pad MFCC if too short for delta computation
        if mfcc.shape[1] < min_frames:
            pad_width = min_frames - mfcc.shape[1]
            mfcc = np.pad(mfcc, ((0, 0), (0, pad_width)), mode='reflect')

        # Compute first and second order deltas with width=min_frames
        delta = librosa.feature.delta(mfcc, width=min_frames)
        delta2 = librosa.feature.delta(mfcc, order=2, width=min_frames)

        # Aggregate features (mean along time axis)
        combined = np.hstack((
            np.mean(mfcc.T, axis=0),
            np.mean(delta.T, axis=0),
            np.mean(delta2.T, axis=0)
        ))

        return combined

    except Exception as e:
        print(f"Error processing {file_path}: {e}")
        return None
```

```python
import pandas as pd
import os

def get_category_from_csv(csv_path, file_path):
    """
    Search for file_path's filename in CSV and return its category.
    """
    metadata_df = pd.read_csv(csv_path)
    filename = os.path.basename(file_path)

    match = metadata_df.loc[metadata_df['filename'] == filename]
    if not match.empty:
        return match.iloc[0]['category']
    else:
        return "unknown"
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,Activation,Flatten
from tensorflow.keras.optimizers import Adam
from sklearn import metrics
from tensorflow.keras.layers import Input
```

```python
num_labels = len(np.unique(y_train))   # if y_train contains class labels
```

```python
# model=Sequential()
# ###first layer
# model.add(Dense(100,input_shape=(40,)))
# model.add(Activation('relu'))
# model.add(Dropout(0.5))
# ###second layer
# model.add(Dense(200))
# model.add(Activation('relu'))
# model.add(Dropout(0.5))
# ###third layer
# model.add(Dense(100))
# model.add(Activation('relu'))
# model.add(Dropout(0.5))

# ###final layer
# model.add(Dense(num_labels))
# model.add(Activation('softmax'))

model = Sequential([
    Input(shape=(40,)),
    Dense(100, activation='relu'),
    Dropout(0.5),
    Dense(200, activation='relu'),
    Dropout(0.5),
    Dense(100, activation='relu'),
    Dropout(0.5),
    Dense(num_labels, activation='softmax')
])

model.compile(optimizer=Adam(learning_rate=0.001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
import os
from datetime import datetime
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

# --- Encode labels ---
le = LabelEncoder()
y_train_int = le.fit_transform(y_train)   # convert strings -> integers
y_test_int = le.transform(y_test)

num_labels = len(le.classes_)             # number of unique classes
y_train_cat = to_categorical(y_train_int, num_classes=num_labels)
y_test_cat = to_categorical(y_test_int, num_classes=num_labels)

# --- Training parameters ---
num_epochs = 250
num_batch_size = 40

# --- Ensure save directory exists ---
save_dir = r"D:\TECH\New folder iiser\New folder\UrbanSound8K"
os.makedirs(save_dir, exist_ok=True)

# --- Callbacks ---
checkpointer = ModelCheckpoint(
    filepath=os.path.join(save_dir, "audio_classification3.keras"),
    monitor='val_accuracy',
    verbose=1,
    save_best_only=True
)

earlystop = EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True
)

# --- Train ---
start = datetime.now()

history = model.fit(X_train, y_train_cat,
                    batch_size=num_batch_size,
                    epochs=num_epochs,
                    validation_data=(X_test, y_test_cat),
                    callbacks=[checkpointer, earlystop],
                    verbose=1)

duration = datetime.now() - start
print("Training completed in time:", duration)
```
